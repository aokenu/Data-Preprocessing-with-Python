<h2>Step 6: Outliers Treatments<h2/>

<p>In data analysis, an outlier is an observation or data point that significantly deviates from the other observations in a dataset. Outliers can be unusually high or low values and can result from variability in the data,<br> measurement errors, or other anomalies. Identifying and handling outliers is a crucial step in data preprocessing, as they can have a substantial impact on statistical analyses and machine learning models.<p/>

<p>Using the IQR method to detect outliers in the dataset<p/>

```python
def wisker(col):
    q1, q3 = np.percentile(col, [25, 75])  # Calculate the first and third quartiles
    iqr = q3 - q1  # Calculate the interquartile range
    lw = q1 - 1.5 * iqr  # Calculate the lower whisker
    uw = q3 + 1.5 * iqr  # Calculate the upper whisker
    return lw, uw  # Return the lower and upper whiskers
```

```python
wisker(df['GDP'])
```

```python
# filling the outliers with the values of the lower wisker and upper wisker

for i in ['GDP', 'Total expenditure', ' thinness  1-19 years', ' thinness 5-9 years']:
    lw,uw = wisker(df[i])
    df[i]=np.where(df[i] < lw,lw,df[i])
    df[i]=np.where(df[i] > uw,uw,df[i])
```

```python
# ploting a boxplot to visualize the values of the affected columns to check that the outlier has been treated successfully

for i in ['GDP', 'Total expenditure', ' thinness  1-19 years', ' thinness 5-9 years']:
    sns.boxplot(df[i])
    plt.show()
```
